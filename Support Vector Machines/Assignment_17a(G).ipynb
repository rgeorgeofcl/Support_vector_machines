##**FEARURE ENGINEERING**

```python
le=LabelEncoder()
columns=['month','day','size_category']
for i in columns:
  data[i]=le.fit_transform(data[i])
data
```

```python
def norm_func(i):
    x=(i-i.min())/(i.max()-i.min())
    return(x)
```

```python
data=norm_func(data)
data
```

```python
x=data.iloc[:,:-1]
x
```

```python
y=data.iloc[:,-1]
y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=200)
```

## GridSearchCV

```python
model=SVC()
param_grid={'kernel':['rbf','sigmoid','poly','linear'],'C':[0.001,0.01,0.05],'gamma':[0.01,1,5,6,10,17,0.01]}
grid_search=GridSearchCV(model,param_grid,cv=10)
```

```python
grid_search.fit(x_train,y_train)
```

```python
print("Best params :",grid_search.best_params_, "  " ,"Best score :" , grid_search.best_score_)
```

# Model Fitting

```python
model1=SVC(kernel='poly',C=0.05,gamma=17)
model1.fit(x_train,y_train)
```

```python
y_pred_train=model1.predict(x_train)
```

```python
y_pred_test=model1.predict(x_test)
```

```python
accuracy_train=accuracy_score(y_train,y_pred_train)
accuracy_train
```

```python
accuracy_test=accuracy_score(y_test,y_pred_test)
accuracy_test
```
